```python
#Import scikit-learn dataset library
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

#Load dataset
breast_cancer_data = datasets.load_breast_cancer()
```

```python
# print the names of the features
print("features - ", breast_cancer_data.feature_names, sep="\n")

# print the label type 
print("\n Labels - ",breast_cancer_data.target_names)

# print data(feature)shape
print("Number of examples are ",breast_cancer_data.data.shape)
print("\nData\n")
print(breast_cancer_data.data)
```

```python
#import the necessary module
from sklearn.model_selection import train_test_split

#split data set into train and test sets
X_train,X_test,Y_train,Y_test = train_test_split(breast_cancer_data.data, breast_cancer_data.target, test_size = 0.4, random_state = 64)
```

```python
#Create a Decision Tree Classifier (using Gini)
dtc = DecisionTreeClassifier(criterion = "gini",max_leaf_nodes=64,random_state=64)

#Train the model using the training sets
dtc.fit(X_train,Y_train)
```

```python
# Predict the classes of test data
pred = dtc.predict(X_test)
print("prediction:",pred)

#print(test_pred.dtype)
```

```python
# Model Accuracy, how often is the classifier correct?
from sklearn import metrics
Accuracy = metrics.accuracy_score(Y_test, pred)
print("Accuracy:",Accuracy)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(dtc,out_file='bcw_tree.dot',feature_names=list(breast_cancer_data.feature_names),
               class_names=list(breast_cancer_data.target_names), filled=True)

# Convert to png
from subprocess import call
call(['dot', '-Tpng', 'bcw_tree.dot', '-o', 'bcw_tree.png', '-Gdpi=600'])
```

```python
# Display in python
import matplotlib.pyplot as plt
plt.figure(figsize = (14, 18))
plt.imshow(plt.imread('bcw_tree.png'))
plt.axis('off')
plt.show()
```
